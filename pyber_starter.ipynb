```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)

#city_data_df.head()
#ride_data_df.head()

# Combine the data into a single dataset
pyber_combined = pd.merge(ride_data_df, city_data_df, how="left", on=["city"])

# Display the data table for preview
pyber_combined
```

## Bubble Plot of Ride Sharing Data

```python
# Group rider by city type
grouped_rcount_type = pyber_combined.groupby(["type","city"])
grouped_rcount_type.head(20)
# Avg fare ($) Per City, Total Number of Rides Per City, Total Number of Drivers Per City
avg_fare = grouped_rcount_type["fare"].mean()
avg_fare.head(20)
rider_count = grouped_rcount_type["ride_id"].count()
rider_count.head(20)
driver_count = grouped_rcount_type["type","city","driver_count"].head(1).set_index(['type','city']).sort_values(["type","city"],ascending = True)*15
driver_count.head(20)
avg_rural=avg_fare["Rural"]
avg_suburban=avg_fare["Suburban"]
avg_urban=avg_fare["Urban"]
ride_rural=rider_count["Rural"]
ride_suburban=rider_count["Suburban"]
ride_urban=rider_count["Urban"]
index = rider_count.index.levels[0]


fig.plt = plt.subplots(figsize = (9,7))
color = ["gold","Skyblue","coral"]
for i in range(len(index)):
    plt.scatter(rider_count[index[i]],avg_fare[index[i]],s=driver_count,marker = "o",facecolor = color[i],edgecolor="k", alpha=.90)
#driver_count

plt.xlabel("Total Number of Rides (Per City)", fontsize = 15) ; plt.ylabel("Average Fare ($)", fontsize = 15)
plt.xticks(size = 12) ; plt.yticks(size = 12)
plt.title("Pyber Ride Sharing Data (2016)", fontsize = 18)
plt.legend(fontsize = "small", loc = "best", fancybox=True)
#plt.show();



# Tells matplotlib that we want to make a scatter plot
# The size of each point on our plot is determined by their x value
#plt.scatter(x_axis, data, marker="o", facecolors="red", edgecolors="black",
            #s=x_axis, alpha=0.75)



#unstacking data
#outcome_drive = drive_mean["fare"].unstack()
#outcome_drive.head()
```
